```python
import pandas as pd
import json

df = pd.read_csv("GraphTrace-2017.csv")

# print( df_hotspots_2017 )

# Convert each POINT to "lat, lon" string
coords_list = df["geometry"].apply(lambda x: f"{x.split()[1]}, {x.split()[2].replace(')','')}").to_list()

# Save to a JSON file
with open("coordinates.json", "w") as f:
    json.dump(coords_list, f, indent=2)

# print(json_array)
```
