```python
import pandas as pd
import glob
import os

files = glob.glob("hotspots/output/*.json")

dataframes = {
    os.path.basename(f).replace(".json", "").replace("output_", ""): pd.read_json(f)
    for f in files
}
```

```python
print(dataframes["9_all_buildingwalk_uniform"])
```

```python
import folium
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

palette = plt.colormaps["tab20"].colors
num_colors = len(palette)

for name, df in dataframes.items():
    
    if df.empty:
        continue

    df = df.copy()
    df = df.dropna(subset=["center", "coverage_area"])

    df["longitude"] = df["center"].apply(
        lambda x: x["coordinates"][0] if isinstance(x, dict) else None
    )
    df["latitude"] = df["center"].apply(
        lambda x: x["coordinates"][1] if isinstance(x, dict) else None
    )

    df = df.dropna(subset=["longitude", "latitude"])

    # sort by score
    df = df.sort_values("best_score", ascending=False).reset_index(drop=True)

    # map center
    map_center = [df["latitude"].mean(), df["longitude"].mean()]
    m = folium.Map(
        location=map_center,
        zoom_start=13,
        tiles="OpenStreetMap",
        control_scale=True
    )

    for i, row in df.iterrows():
        rank_label = row["rank"]
        lat = row["latitude"]   # ✅ FIX
        lon = row["longitude"]  # ✅ FIX

        if row["coverage_area"] is None or len(row["coverage_area"]) == 0:
            continue

        polygon = row["coverage_area"][0]

        if polygon is None or len(polygon) == 0:
            continue

        polygon_coords = [(p[1], p[0]) for p in polygon]

        color = mcolors.to_hex(palette[i % num_colors])

        # 🎨 polygon
        folium.Polygon(
            locations=polygon_coords,
            color=color,
            weight=2,
            fill=True,
            fill_color=color,
            fill_opacity=0.4,
            tooltip=f"Score: {row['best_score']} | Crimes: {row['seen_crimes']}"
        ).add_to(m)

        # 🔴 center point
        folium.CircleMarker(
            location=[lat, lon],
            radius=5,
            color="red",
            fill=True,
            fill_color="red",
            fill_opacity=1
        ).add_to(m)

        folium.Marker(
            location=[lat, lon],
            icon=folium.DivIcon(
                html=f'''
                <div style="
                    font-size:16px;
                    font-weight:bold;
                    color:white;
                    text-align:center;
                    text-shadow: 
                        1px 1px 2px black,
                        -1px -1px 2px black,
                        1px -1px 2px black,
                        -1px 1px 2px black;
                ">
                    {rank_label}
                </div>
                '''
            )
        ).add_to(m)

    # save map
    m.save(f"hotspots/output/html/{name}_map.html")
```
