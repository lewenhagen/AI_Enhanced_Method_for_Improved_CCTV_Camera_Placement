```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
import numpy as np
from scipy.stats import friedmanchisquare

exp_json = "./experiment-20260218.json"

#methods = ["bruteforce", "dfs", "hillclimb", "buildingwalk"]

df = pd.read_json(exp_json)
```

```python
df["block"] = df["activation_function"] + "_" + df["radius"].astype(str)
```

```python
pivot_df = df.pivot_table(
    index="block",
    columns="method",
    values="pai"
)
pivot_df = pivot_df.dropna()
display(pivot_df)
```

```python
stat, p = friedmanchisquare(
    pivot_df["bruteforce"],
    pivot_df["dfs"],
    pivot_df["hillclimb"],
    pivot_df["buildingwalk"]
)

print("Friedman statistic:", stat)
print("p-value:", p)
```

```python
# Nemenyi post hoc test
import scikit_posthocs as sp

posthoc = sp.posthoc_nemenyi_friedman(pivot_df)
print(posthoc)
```

```python
# Kendalls W

k = 4  
n = len(pivot_df)

W = stat / (n * (k - 1))
print("Kendall's W:", W)
```

```python
# Average Rank
pivot_df.rank(axis=1, ascending=False).mean()
```
